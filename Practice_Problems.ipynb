```python
#shell game problem 1
shells = ["yellow","pink","blue"]
n = int(input("enter the number of swaps: "))
swaps = []
for i in range(n):
  swaps.append([int(x) - 1 for x in input().split()])
yellow_points = 0
pink_points = 0
blue_points = 0
for a, b, g in swaps:
  shells[a], shells[b] = shells[b], shells[a]
  if shells[g] == "yellow":
    yellow_points += 1
  elif shells[g] == "pink":
    pink_points += 1
  else:
    blue_points += 1
result = max(yellow_points, pink_points, blue_points)
print(result)
```

```python
#1873G ABBC or BACB
empty_list = []
t = int(input())
for i in range(t):
  num_of_coins = 0
  s = input().strip()
  while "AB" in s or "BA" in s:
    if s[0] == "A":
      if "AB" in s:
        s = s.replace("AB", "BC", 1)
        num_of_coins += 1
      elif "BA" in s:
        s = s.replace("BA","CB",1)
        num_of_coins += 1
      else:
        break
    else:
      if "BA" in s:
        s = s.replace("BA", "CB", 1)
        num_of_coins += 1
      elif "AB" in s:
        s = s.replace("AB", "BC", 1)
      else:
        break
  empty_list.append(num_of_coins)
for coin in empty_list:
  print(coin)
```

```python
#problem 1 diamond collector
total_diamonds = []
diamond_weights = []
max_counter = 0
n, k = [int(x) for x in input().split()]
for i in range(n):
  weights = int(input())
  diamond_weights.append(weights)
diamond_weights.sort()
for i in range(n):
  simple_counter = 0
  for j in range(i, n):
    difference = abs(diamond_weights[i] - diamond_weights[j])
    if difference <= k:
      simple_counter += 1
  if simple_counter > max_counter:
    max_counter = simple_counter
print(max_counter)
```

```python
#A maximum distance
n = int(input())
x = list(map(int, input().split()))
y = list(map(int, input().split()))
maximum_distance = []
for i in range(n):
  for j in range(n):
    distance = (((x[i] - x[j])**2) + ((y[i] - y[j])**2))
    maximum_distance.append(distance)
    result = max(maximum_distance)
print(result)
```

```python
#alice and bob
alice_x, alice_y = 0,0
r, s = map(int, input().split())
m, n, p, q = map(int, input().split())
seconds = 0
while (alice_x < r) and (alice_y < s):
  alice_x = alice_x + m
  alice_y = alice_y + n
  r = r - p
  s = s - q
  seconds += 1
  if alice_x == r and alice_y == s:
    print(seconds)
  else:
    print(-1)
```

```python
#problem 3, why did the cows cross the road III
times = []
n = int(input())
for i in range(n):
  arrival_time, durration_time = map(int, input().split())
  times.append((arrival_time, durration_time))
ad_timings = 0
times.sort()
for arrival_time, durration_time in times:
  if arrival_time > ad_timings:
    ad_timings = arrival_time
    ad_timings += durration_time
  elif arrival_time <= ad_timings:
    ad_timings += durration_time
print(ad_timings)
```

```python
#cses, apartments
n, m, k = map(int, input().split())
a = list(map(int, input().split()))
b = list(map(int, input().split()))
a.sort()
b.sort()
i = 0
j = 0
counter = 0
while i<n and j<m:
  if b[j] < a[i] - k:
    j += 1
  elif b[j] > a[i] + k:
    i += 1
  else:
    counter +=1
    j += 1
    i += 1
print(counter)
```

```python
#cses, ferris wheel
n, x = map(int, input().split())
p = list(map(int, input().split()))
p.sort()
i = 0
j = n-1
counter = 0
while i <= j:
  if p[i] + p[j] < x:
    i += 1
    j -= 1
  else:
    j-=1
  counter += 1
print(counter)
```

```python
#voronoi villages ccc'18 s1
n = int(input())
villages = []
for x in range(n):
  v = int(input())
  villages.append(v)
villages.sort()
min_neighborhood = float('inf')
for i in range(1, n-1):
  left_size = (villages[i] + villages[i-1])/2
  right_size = (villages[i+1] + villages[i]) / 2
  total_size = right_size - left_size
  min_neighborhood = min(min_neighborhood, total_size)
print(f"{min_neighborhood:.1f}")
```

```python
#waterbuckets
n = int(input())
c = list(map(int, input().split()))
a = list(map(int, input().split()))
for i in range(n-1):
  giver_bucket = a[i]
  reciving_bucket = a[i+1]
  remaining_capacity = c[i+1] - reciving_bucket
  if giver_bucket > remaining_capacity:
    reciving_bucket = c[i+1]
    giver_bucket -= remaining_capacity
  else:
    reciving_bucket += giver_bucket
    giver_bucket = 0
  a[i] = giver_bucket
  a[i+1] = reciving_bucket
print(*a)
```

```python
#problem 3 mowing the field
n = int(input())
directions = []
ans = float('inf')
dic = {}
for i in range(n):
  d, s = input().split()
  directions.append([d,int(s)])
#print(directions)
x, y = 0,0
dic[(x,y)] = 0
t = 0
for direction,distance in directions:
  for i in range(distance):
    t += 1
    if direction == "W":
      x -= 1
    elif direction == "S":
      y -= 1
    elif direction == "N":
      y += 1
    elif direction == "E":
      x += 1
    if (x,y) in dic:
      gap = (t - dic[(x,y)])
      ans = min(gap, ans)
    dic[(x,y)] = t
if ans == float('inf'):
  print(-1)
else:
  print(ans)
```

```python
# rudolph problem
n, d, h = map(int, input().split())
b_h = list(map(int, input().split()))
areas = ((d*h)/2) * 3
#how to find the overlapping. if next triangle starts before previous triangle ends.
for x in range(n-1):
  top_of_previous_triangle = h + b_h[x]
  if top_of_previous_triangle > b_h[x+1]:
    overlapped_height = top_of_previous_triangle - b_h[x+1]
    small_triangle_area = ((d*overlapped_height**2)/(2*h))
    areas = areas - small_triangle_area
print(areas)
#subtract the little piece of the triangle by the large piece above it
```
